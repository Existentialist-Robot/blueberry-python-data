```python
##########################################################################################################
# Pass the table to a dataframe
##########################################################################################################

conn = sqlite3.connect('fnirs_data.db')
cursor = conn.cursor()

TARGET = table_name
FNIRSDATA = pd.read_sql_query('SELECT * FROM ' + TARGET , conn)
#print(FNIRSDATA)
```

```python
FNIRSDATA.columns
##########################################################################################################
# Remove first 50 rows from data from for initial noise from session data, when starting ambient light adjusts the initial values
##########################################################################################################
FNIRSDATA = FNIRSDATA.iloc[50:]
```

```python
##########################################################################################################
# Start to perform simple statiscal smoothing techniques on the HbO2 data
# Techniques include: Rolling Averages, Standard Deviations
# Standard Smoothing Techiques for Fast Moving Stocks Applied as common practice with realtime sensor data
# https://nbviewer.jupyter.org/urls/alphabench.com/data/pandas_rolling_method.ipynb
##########################################################################################################
import numpy as np
import pandas as pd
import itertools as it

HbO2 = FNIRSDATA['HbO2']
HbO2 = HbO2[~np.isnan(HbO2)]
# print(P1)
# df.plot(x='index', y='P1', kind='line')

#DATA ANALYSIS#
# calculate rolling window of 100 samples or 20 seconds at 5Hz, 20 seconds and 4 seconds
FNIRSDATA['RA_20'] = FNIRSDATA.HbO2.rolling(20).mean().shift()
FNIRSDATA['RA_100'] = FNIRSDATA.HbO2.rolling(100).mean()
FNIRSDATA['RA_20_100'] = FNIRSDATA.RA_20.rolling(100).mean()
FNIRSDATA['RA_20'].head(12)

FNIRSDATA['STD_20'] = FNIRSDATA.HbO2.rolling(20).std().shift()
FNIRSDATA['STD_100'] = FNIRSDATA.HbO2.rolling(100).std()
FNIRSDATA['STD_20_100'] = FNIRSDATA.STD_20.rolling(100).std()
FNIRSDATA['STD_DIFF_20_100'] = FNIRSDATA['STD_20'] - FNIRSDATA['STD_100']
FNIRSDATA['STD_DIFF_MEAN'] = FNIRSDATA.STD_DIFF_20_100.rolling(100).mean()

# #DATA PLOTS AVERAGES#
plt.figure(figsize=(15,10))
plt.grid(True)
plt.plot(FNIRSDATA['HbO2'],label='HBO2')
plt.plot(FNIRSDATA['RA_20'], label='RA 20 samples')
plt.ylabel('Rolling Average 5s')
plt.xlabel('Time')
plt.legend(loc=2)

# #DATA PLOTS AVERAGES#
plt.figure(figsize=(15,10))
plt.grid(True)
plt.plot(FNIRSDATA['HbO2'],label='HBO2')
plt.plot(FNIRSDATA['RA_20_100'], label="RA_20_100 samples")
plt.ylabel('Rolling Average 20s from rolling average of 4 seconds')
plt.xlabel('Time')
plt.legend(loc=2)
```

```python
##########################################################################################################
# Taking the rolling average of the filtered signal
# To amplify change in signal or our case change in blood flow as a represenation of regional activation
# This plot show the rate of change in the rolling average filtered signal
##########################################################################################################

FNIRSDATA["RA_20_100"] = FNIRSDATA["RA_20_100"] + 20
FNIRSDATA["change"] = np.log(FNIRSDATA["RA_20_100"] / FNIRSDATA["RA_20_100"].shift())
plt.ylabel('Rate of Change of HbO2 Rolling Average')
plt.xlabel('Time')
plt.plot(FNIRSDATA.change)
```

```python
##########################################################################################################
# Start to perform simple statiscal smoothing techniques on the HbO2 data
# To further showcase change in blood flow we will explore the volatility of the rate of change by applying standard deviation
# For the data set experiment we will assume that the accessory was properly placed next to the head with no significant lighting changes
# ie. person was sitting at their desk working
##########################################################################################################

FNIRSDATA['Volatility'] = FNIRSDATA.change.rolling(20).std().shift()
# Time is an integer value of samples at a sample rate of 5Hz counting up from start of session

# Smoothen Volatility
FNIRSDATA['Volatility_RollingAvg'] = FNIRSDATA.Volatility.rolling(100).mean()
FNIRSDATA['Volatility_RollingAvg'] = FNIRSDATA.Volatility_RollingAvg * 1000000 #scale by 1,000,000
# print(FNIRSDATA.Volatility_RollingAvg.count)

#create new array of 3346 in length or 10 mins or each minute value is 300 or 60 second steps
WPM = [62.5, 64.3, 65.0, 63.1, 62.3, 64.7, 69.2, 64.0, 55.0, 59.3, 53.0]
WPM_TOTAL = []

for i in range(len(WPM)):
    tempWPM = WPM[i]
    for i in range(300):    
        WPM_TOTAL.append(tempWPM)

#print(len(WPM_TOTAL))    
#print(WPM_TOTAL)
FNIRSDATA['WPM'] = pd.Series(WPM_TOTAL)

plt.grid(True)
# plt.plot(FNIRSDATA['Volatility'],label='Volatility')
plt.plot(FNIRSDATA['Volatility_RollingAvg'], label='Volatility_RollingAvg')
plt.plot(FNIRSDATA['WPM'], label='WPM')
plt.xlabel('Time')
plt.legend(loc=2)

FNIRSDATA.Volatility_RollingAvg.dropna().describe()
```

```python
##########################################################################################################
# Pass the table to a dataframe
##########################################################################################################

conn_wpm = sqlite3.connect('fnirs_data_wpm.db')
cursor = conn_wpm.cursor()

TARGET_WPM = table_name_wpm
FNIRSDATA_WPM = pd.read_sql_query('SELECT * FROM ' + TARGET_WPM , conn_wpm)
```

```python
FNIRSDATA_WPM.columns
##########################################################################################################
# Remove first 50 rows from data from for initial noise from session data, when starting ambient light adjusts the initial values
##########################################################################################################
FNIRSDATA_WPM = FNIRSDATA_WPM.iloc[50:]
```

```python
##########################################################################################################
# Combining and simplifying techniques from above for filtration
##########################################################################################################
import numpy as np
import pandas as pd
import itertools as it

T1 = FNIRSDATA_WPM['T1']
T1 = T1[~np.isnan(T1)]

T2 = FNIRSDATA_WPM['T2']
T2 = T1[~np.isnan(T2)]

T3 = FNIRSDATA_WPM['T3']
T3 = T1[~np.isnan(T3)]

T4 = FNIRSDATA_WPM['T4']
T4 = T1[~np.isnan(T4)]

T5 = FNIRSDATA_WPM['T5']
T5 = T1[~np.isnan(T5)]

#DATA ANALYSIS#
# calculate rolling window of 5, 20 samples, 1 and 5 seconds
FNIRSDATA_WPM['RA_5_T1'] = FNIRSDATA_WPM.T1.rolling(5).mean().shift()
FNIRSDATA_WPM['RA_20_T1'] = FNIRSDATA_WPM.T1.rolling(20).mean()
FNIRSDATA_WPM['RA_5_20_T1'] = FNIRSDATA_WPM.RA_5_T1.rolling(20).mean()

FNIRSDATA_WPM["change_T1"] = np.log(FNIRSDATA_WPM["RA_5_20_T1"] / FNIRSDATA_WPM["RA_5_20_T1"].shift())
FNIRSDATA_WPM['Volatility_T1'] = FNIRSDATA_WPM.change_T1.rolling(5).std().shift()
FNIRSDATA_WPM['Volatility_RollingAvg_T1'] = FNIRSDATA_WPM.Volatility_T1.rolling(20).mean()

FNIRSDATA_WPM['RA_5_T2'] = FNIRSDATA_WPM.T2.rolling(5).mean().shift()
FNIRSDATA_WPM['RA_20_T2'] = FNIRSDATA_WPM.T2.rolling(20).mean()
FNIRSDATA_WPM['RA_5_20_T2'] = FNIRSDATA_WPM.RA_5_T2.rolling(20).mean()

FNIRSDATA_WPM["change_T2"] = np.log(FNIRSDATA_WPM["RA_5_20_T2"] / FNIRSDATA_WPM["RA_5_20_T2"].shift())
FNIRSDATA_WPM['Volatility_T2'] = FNIRSDATA_WPM.change_T2.rolling(5).std().shift()
FNIRSDATA_WPM['Volatility_RollingAvg_T2'] = FNIRSDATA_WPM.Volatility_T2.rolling(20).mean()

FNIRSDATA_WPM['RA_5_T3'] = FNIRSDATA_WPM.T3.rolling(5).mean().shift()
FNIRSDATA_WPM['RA_20_T3'] = FNIRSDATA_WPM.T3.rolling(20).mean()
FNIRSDATA_WPM['RA_5_20_T3'] = FNIRSDATA_WPM.RA_5_T3.rolling(20).mean()

FNIRSDATA_WPM["change_T3"] = np.log(FNIRSDATA_WPM["RA_5_20_T3"] / FNIRSDATA_WPM["RA_5_20_T3"].shift())
FNIRSDATA_WPM['Volatility_T3'] = FNIRSDATA_WPM.change_T3.rolling(5).std().shift()
FNIRSDATA_WPM['Volatility_RollingAvg_T3'] = FNIRSDATA_WPM.Volatility_T3.rolling(20).mean()

FNIRSDATA_WPM['RA_5_T4'] = FNIRSDATA_WPM.T4.rolling(5).mean().shift()
FNIRSDATA_WPM['RA_20_T4'] = FNIRSDATA_WPM.T4.rolling(20).mean()
FNIRSDATA_WPM['RA_5_20_T4'] = FNIRSDATA_WPM.RA_5_T4.rolling(20).mean()

FNIRSDATA_WPM["change_T4"] = np.log(FNIRSDATA_WPM["RA_5_20_T4"] / FNIRSDATA_WPM["RA_5_20_T4"].shift())
FNIRSDATA_WPM['Volatility_T4'] = FNIRSDATA_WPM.change_T4.rolling(5).std().shift()
FNIRSDATA_WPM['Volatility_RollingAvg_T4'] = FNIRSDATA_WPM.Volatility_T4.rolling(20).mean()

FNIRSDATA_WPM['RA_5_T5'] = FNIRSDATA_WPM.T4.rolling(5).mean().shift()
FNIRSDATA_WPM['RA_20_T5'] = FNIRSDATA_WPM.T4.rolling(20).mean()
FNIRSDATA_WPM['RA_5_20_T5'] = FNIRSDATA_WPM.RA_5_T5.rolling(20).mean()

FNIRSDATA_WPM["change_T5"] = np.log(FNIRSDATA_WPM["RA_5_20_T5"] / FNIRSDATA_WPM["RA_5_20_T5"].shift())
FNIRSDATA_WPM['Volatility_T5'] = FNIRSDATA_WPM.change_T5.rolling(5).std().shift()
FNIRSDATA_WPM['Volatility_RollingAvg_T5'] = FNIRSDATA_WPM.Volatility_T5.rolling(20).mean()

# plt.grid(True)
# # plt.plot(FNIRSDATA_WPM['T1'],label='T1')
# plt.plot(FNIRSDATA_WPM['Volatility_T1'],label='Volatility_T1')
# plt.plot(FNIRSDATA_WPM['Volatility_RollingAvg_T1'], label='Volatility_RollingAvg_T1')
# plt.xlabel('Time')
# plt.legend(loc=2)

plt.grid(True)
# plt.plot(FNIRSDATA_WPM['T1'],label='T1')
plt.plot(FNIRSDATA_WPM['Volatility_RollingAvg_T1'], label='T1 54.3 wpm, acc 96.7') #54.3 wpm, accuracy 96.7
plt.plot(FNIRSDATA_WPM['Volatility_RollingAvg_T2'], label='T2 60.2 wpm, acc 96.9') #60.2 wpm, accuracy 96.9
plt.plot(FNIRSDATA_WPM['Volatility_RollingAvg_T3'], label='T3 57.9 wpm, acc 95.8') #57.9 wpm, accuracy 95.8
plt.plot(FNIRSDATA_WPM['Volatility_RollingAvg_T4'], label='T4 56 wpm, acc 95') #56 wpm, accuracy 95
plt.plot(FNIRSDATA_WPM['Volatility_RollingAvg_T5'], label='T5 59.5 wpm, acc 94.5') #59.5 wpm, accuracy 94.5

plt.xlabel('Time')
plt.legend(loc=1)

FNIRSDATA_WPM.Volatility_RollingAvg_T1.dropna().describe()
FNIRSDATA_WPM.Volatility_RollingAvg_T2.dropna().describe()
FNIRSDATA_WPM.Volatility_RollingAvg_T3.dropna().describe()
FNIRSDATA_WPM.Volatility_RollingAvg_T4.dropna().describe()
FNIRSDATA_WPM.Volatility_RollingAvg_T5.dropna().describe()
```
